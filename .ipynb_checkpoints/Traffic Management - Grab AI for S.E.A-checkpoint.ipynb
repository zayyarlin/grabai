## Baseline model
Predict that demand at T and location L is same as mean demand at T and L from the look back period

```python
def score(input, start_day = 0, t_start = '0:0', abs_time_start = None, look_back = 14):
    # Define relative time 
    # 00:00 -> 0
    # 00:15 -> 1 etc
    # ...
    # 23:45 -> 95
    time_cache={}
    def time_str_to_rel_time(time_str):
        if time_str in time_cache:
            return time_cache[time_str]
        else:
            hour = time_str.split(':')[0]
            minute = time_str.split(':')[1]
            fifteen_min_bin = {0 : 0, 15: 1, 30: 2, 45: 3}
            result = (int(hour) * 4) + fifteen_min_bin[int(minute)]
            time_cache[time_str] = result 
            return result
    
    input['rel_time'] = input['timestamp'].apply(time_str_to_rel_time)
    input['abs_time'] = ((input['day']-1)*96) + input['rel_time']
    
    if abs_time_start is None:
        t = (start_day - 1) * 96 + time_str_to_rel_time(t_start)
    else:
        t = abs_time_start 
    training_set_start = t - (look_back * 96)
    
    pred_set = input.query(f" {t+1} < abs_time <= {t+5} ").copy()
    
    # train the data for look back period
    train_set = input.query(f" {training_set_start} < abs_time <= {t} ")

    # cluster the days into three types, peak, non-peak, mid-week
    grouped = train_set.groupby(['geohash6', 'rel_time'])['demand'].mean().unstack(level=-1, fill_value=0.0)
    
    def predict(geohash, rel_time):
        try:
            return grouped.loc[geohash][rel_time]
        except KeyError:
            return 0.0

    pred_set['pred_demand'] = pred_set.apply(lambda x: predict(x['geohash6'], x['rel_time']), axis=1)
    return pred_set
    
    
def RMSE(df):
    return ((df['demand'] - df['pred_demand']) ** 2).mean() ** 0.5
```

```python
# Look at the RMSE scores across days
start_day = 14
end_day = 28

rmse_scores = []
for day in range(start_day, end_day):
    for hour in range(9,24,3): # 9 am, 12 noon, 3 pm, 6 pm, 9 pm
        start_time = (day * 96) + (hour * 4)
        rmse = RMSE(score(demand_df, abs_time_start = start_time, look_back=14))
        rmse_scores.append((day, hour, start_time, rmse))
```

```python
rmse_df = pd.DataFrame(rmse_scores, columns = ['day', 'hour', 'abs_time', 'rmse'])
```

We see that the cluster labels (generally) distingish between peak days and trough days. We shall attempt to modify the predict function to take advantage of this.

```python
def score_with_daytype(gmm, input, start_day = 0, t_start = '0:0', abs_time_start = None, look_back = 14):
    time_cache={}
    def time_str_to_rel_time(time_str):
        if time_str in time_cache:
            return time_cache[time_str]
        else:
            hour = time_str.split(':')[0]
            minute = time_str.split(':')[1]
            fifteen_min_bin = {0 : 0, 15: 1, 30: 2, 45: 3}
            result = (int(hour) * 4) + fifteen_min_bin[int(minute)]
            time_cache[time_str] = result 
            return result
        
    input['rel_time'] = input['timestamp'].apply(time_str_to_rel_time)
    input['abs_time'] = ((input['day']-1)*96) + input['rel_time']
    
    if abs_time_start is None:
        t = (start_day - 1) * 96 + time_str_to_rel_time(t_start)
    else:
        t = abs_time_start 
    training_set_start = t - (look_back * 96)
    
    pred_set = input.query(f" {t+1} < abs_time <= {t+5} ").copy()
    
    # train the data for look back period
    train_set = input.query(f" {training_set_start} < abs_time <= {t} ").copy()
    
    # get the demand data on the same day as t+1, ...., t+5
    # day number of pred_set
    pred_set_day = pred_set['day'].min()
    max_time_on_pred_day = train_set.query(f"day == {pred_set_day}")['rel_time'].max()
    
    if (max_time_on_pred_day >= 36):
        # print("Enough info for day_type_label")
        pred_day_input_for_day_label = train_set.query(f"day == {pred_set_day} & rel_time < 36")
    

        # cluster the days into three types, peak, non-peak, mid-week
        # Take only rel_time < 36 (< 9 am) so that we can use the morning info
        day_cluster = train_set.query("rel_time < 36").groupby(['day','rel_time'])['demand'].sum().unstack(level=-1, fill_value=0.0)
                
        #pca = PCA(n_components=2).fit(day_cluster_delta)
        #day_cluster_reduced = pd.DataFrame(pca.transform(day_cluster_delta), index=day_cluster_delta.index)

        # gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=42).fit(day_cluster_delta)
        labels = gmm.predict(day_cluster)
        # print(labels)
        day_cluster['clabel'] = labels
        
        memoized = {} # otherwise too slow
        def find_day_type(day):
            if day in memoized:
                return memoized[day]
            else:
                result = -2.0
                try:
                    result = day_cluster.loc[day]['clabel']
                except KeyError:
                    pass # Couldn't find a cluster label due to data being cut short
                memoized[day] = result
                return result

        # label the training set with peak/non-peak
        train_set['day_type'] = train_set.apply(lambda x: find_day_type(x['day']), axis=1)
        grouped = train_set.groupby(['geohash6', 'day_type', 'rel_time'])['demand'].mean().unstack(level=-1, fill_value=0.0)
        
        
        # label the pred set
        pred_day = pred_day_input_for_day_label.groupby(['day','rel_time'])['demand'].sum().unstack(level=-1, fill_value=0.0)

        #pred_day_reduced = pd.DataFrame(pca.transform(pred_day_delta), index=pred_day_delta.index)
        #print(pred_day_reduced)
        #print(gmm.predict(pred_day_reduced))
        pred_day_type_label = gmm.predict(pred_day)[0]
        # print("pred daytype label: ", pred_day_type_label)
        pred_set['day_type'] = pred_day_type_label
    else:
        # print("Not enough info for day_type label")
        pred_set['day_type'] = -1
        grouped = train_set.groupby(['geohash6', 'rel_time'])['demand'].mean().unstack(level=-1, fill_value=0.0)
        
    def predict(geohash, rel_time, day_type):
        try:
            if day_type == -1: # No day_type_set
                return grouped.loc[geohash][rel_time]
            else:
                return grouped.loc[geohash][rel_time][day_type]
        except KeyError:
            return 0.0

    pred_set['pred_demand'] = pred_set.apply(lambda x: predict(x['geohash6'], x['rel_time'], x['day_type']), axis=1)
    return pred_set
    
    
def RMSE(df):
    return ((df['demand'] - df['pred_demand']) ** 2).mean() ** 0.5
```

```python
# Look at the RMSE scores across days
start_day = 14
end_day = 28

rmse_scores_w_daytype = []
for day in range(start_day, end_day):
    print("day: ", day)
    for hour in range(9,24,3): # 9 am, 12 noon, 3 pm, 6 pm, 9 pm
        start_time = (day * 96) + (hour * 4)
        rmse = RMSE(score_with_daytype(gmm, demand_df, abs_time_start = start_time, look_back=14))
        rmse_scores_w_daytype.append((day, hour, start_time, rmse))
```

## Final Model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from joblib import dump, load

def time_str_to_rel_time(time_str):
    hour = time_str.split(':')[0]
    minute = time_str.split(':')[1]
    
    fifteen_min_bin = {0 : 0, 15: 1, 30: 2, 45: 3}
    
    return (int(hour) * 4) + fifteen_min_bin[int(minute)]

def final_score(gmm, input, start_day = 0, t_start = '0:0', abs_time_start = None, look_back = 14):
    time_cache={}
    def time_str_to_rel_time(time_str):
        if time_str in time_cache:
            return time_cache[time_str]
        else:
            hour = time_str.split(':')[0]
            minute = time_str.split(':')[1]
            fifteen_min_bin = {0 : 0, 15: 1, 30: 2, 45: 3}
            result = (int(hour) * 4) + fifteen_min_bin[int(minute)]
            time_cache[time_str] = result 
            return result
        
    input['rel_time'] = input['timestamp'].apply(time_str_to_rel_time)
    input['abs_time'] = ((input['day']-1)*96) + input['rel_time']
    
    if abs_time_start is None:
        t = (start_day - 1) * 96 + time_str_to_rel_time(t_start)
    else:
        t = abs_time_start 
    training_set_start = t - (look_back * 96)
    
    pred_set = input.query(f" {t+1} < abs_time <= {t+5} ").copy()
    
    # train the data for look back period
    train_set = input.query(f" {training_set_start} < abs_time <= {t} ").copy()
    
    # get the demand data on the same day as t+1, ...., t+5
    # day number of pred_set
    pred_set_day = pred_set['day'].min()
    max_time_on_pred_day = train_set.query(f"day == {pred_set_day}")['rel_time'].max()
    
    if (max_time_on_pred_day >= 36):
        # print("Enough info for day_type_label")
        pred_day_input_for_day_label = train_set.query(f"day == {pred_set_day} & rel_time < 36")
    

        # cluster the days into three types, peak, non-peak, mid-week
        # Take only rel_time < 36 (< 9 am) so that we can use the morning info
        day_cluster = train_set.query("rel_time < 36").groupby(['day','rel_time'])['demand'].sum().unstack(level=-1, fill_value=0.0)
                
        #pca = PCA(n_components=2).fit(day_cluster_delta)
        #day_cluster_reduced = pd.DataFrame(pca.transform(day_cluster_delta), index=day_cluster_delta.index)

        # gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=42).fit(day_cluster_delta)
        labels = gmm.predict(day_cluster)
        # print(labels)
        day_cluster['clabel'] = labels
        
        memoized = {} # otherwise too slow
        def find_day_type(day):
            if day in memoized:
                return memoized[day]
            else:
                result = -2.0
                try:
                    result = day_cluster.loc[day]['clabel']
                except KeyError:
                    pass # Couldn't find a cluster label due to data being cut short
                memoized[day] = result
                return result

        # label the training set with peak/non-peak
        train_set['day_type'] = train_set.apply(lambda x: find_day_type(x['day']), axis=1)
        grouped = train_set.groupby(['geohash6', 'day_type', 'rel_time'])['demand'].mean().unstack(level=-1, fill_value=0.0)
        
        
        # label the pred set
        pred_day = pred_day_input_for_day_label.groupby(['day','rel_time'])['demand'].sum().unstack(level=-1, fill_value=0.0)

        #pred_day_reduced = pd.DataFrame(pca.transform(pred_day_delta), index=pred_day_delta.index)
        #print(pred_day_reduced)
        #print(gmm.predict(pred_day_reduced))
        pred_day_type_label = gmm.predict(pred_day)[0]
        # print("pred daytype label: ", pred_day_type_label)
        pred_set['day_type'] = pred_day_type_label
    else:
        # print("Not enough info for day_type label")
        pred_set['day_type'] = -1
        grouped = train_set.groupby(['geohash6', 'rel_time'])['demand'].mean().unstack(level=-1, fill_value=0.0)
        
    def predict(geohash, rel_time, day_type):
        try:
            if day_type == -1: # No day_type_set
                return grouped.loc[geohash][rel_time]
            else:
                return grouped.loc[geohash][rel_time][day_type]
        except KeyError:
            return 0.0

    pred_set['pred_demand'] = pred_set.apply(lambda x: predict(x['geohash6'], x['rel_time'], x['day_type']), axis=1)
    return pred_set
    
    
def RMSE(df):
    return ((df['demand'] - df['pred_demand']) ** 2).mean() ** 0.5
```

## Future work

Instead of using individual geohash, we can also cluster them to reduce the degree-of-freedom in the data set.
This will allow us to use more sophisticated models (e.g. Hidden markov model with GaussianMixture emissions).
I did some exploratory work, but weren't able to finish on time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import geohash2
```

```python
demand_df = pd.read_csv('data/training.csv')
demand_df['latlon'] = demand_df['geohash6'].apply(lambda g: geohash2.decode_exactly(g))
demand_df['x'] = demand_df['latlon'].apply(lambda x: x[0])
demand_df['y'] = demand_df['latlon'].apply(lambda x: x[1])
demand_df['rel_time'] = demand_df['timestamp'].apply(time_str_to_rel_time)
demand_df['abs_time'] = ((demand_df['day']-1)*96) + demand_df['rel_time']
```

```python
cluster_df = demand_df.groupby(['geohash6', 'rel_time'])['demand'].mean().unstack(level=-1, fill_value=0.0)
```

```python
# Cluster with TSNE dimensionality reduction
def cluster_dim_red(Y, components, dim_red_plot=False):
    X = pd.DataFrame(TSNE(n_components=2, perplexity=5).fit_transform(Y), index=Y.index)
    gmm = GaussianMixture(n_components=components, covariance_type='full', random_state=0).fit(X)
    labels = gmm.predict(X)
    labelled = pd.concat([pd.Series(X.index.values), pd.Series(labels),X[0].reset_index(drop=True), X[1].reset_index(drop=True)], axis=1)
    labelled.columns = ['geohash6', 'c_label', 'dim_x', 'dim_y']
    labelled['latlon'] = labelled['geohash6'].apply(lambda g: geohash2.decode_exactly(g))
    labelled['x'] = labelled['latlon'].apply(lambda x: x[0])
    labelled['y'] = labelled['latlon'].apply(lambda x: x[1])
    plt.figure(figsize=(10,10))
    if dim_red_plot:
        plt.scatter('dim_x', 'dim_y', data=labelled, c=labelled.c_label, cmap='tab10')
    else:
        plt.scatter('x', 'y', data=labelled, c=labelled.c_label, cmap='tab10')
```

```python
cluster_dim_red(cluster_df, 8)
```

We see clusters of geohashes that can be modeled as single entities

```python
cluster_dim_red(cluster_df, 8, dim_red_plot=True)
```
